```python
catatan = []  #struktur data untuk menyimpan catatan

def tambah(): #fungsi untuk menambahakan catatan
  judul = input("masukan judul catatan: ")  #perintah untuk memasukan judul
  isi = input("masukan isi catatan: ")      #perintah untuk memasukan isi konten
  note = {"Judul": judul, "Isi": isi}       #digunakan untuk menyimpan seluruh catatan
  catatan.append(note)
  print("Catatan ditambah")

def lihat():  #Fungsi untuk melihat catatan yang sudah dibuat
  if len(catatan) == 0: #fungsi len untuk menghitung jumlah catatan, apabila tidak ada catatan maka
    print("catatan masih kosong")   #akan mencetak tulisan disamping
    print("coba buat catatan baru")
  else:
    print("Daftar catatan anda:") #Namun jika ada maka, akan menampilkan seperti ini
    for note in catatan:   #for akan mengakses note dalam catatan
      print("# Judul: ", note["Judul"]) #lalu akan mencetak judul, diambil dari note
      print("# Isi", note["Isi"]) #dan juga akan mencetak isi, diambil dari note
      print("")

def edit(): #fungsi ini digunakan untuk mengedit catatan
  if len(catatan) == 0: #Menghitung catatan, apabila kosong maka
    print("Tidak ada catatan yang tersedia")  #Akan mencetak ini
    return

  lihat()  #Fungsi lihat
  pilihan = int(input("Pilih mana catatan yang ingin diedit: "))  #variable pilihan berisi inputan converting dari string ke integer
  if pilihan < 1 or pilihan > len(catatan): #jika pilihan memasukan angka yang valid6 maka diarahkan ke edit, jika tidak maka
     print("Nomor catatan tidak ada")  #akan muncul tulisan cetak seperti disamping dan
     return  #kembali ke pilihan

  note = catatan[pilihan - 1]
  print("Catatan yang akan diedit: ")
  print("- Judul: ", note["Judul"])
  print("- Isi: ", note["Isi"])
  print("")

  judul_baru = input("masukan judul baru (kosongkan jika tidak ingin diubah)")
  isi_baru = input("masukan konten baru, atau kosongkan jika tidak ingin diubah")

  if judul_baru != "":
    note["Judul"] = judul_baru
  if isi_baru != "":
    note["Isi"] = isi_baru

  print("Catatan telah diubah dan diperbarui")

def hapus():
  if len(catatan) == 0:
    print("tidak ada catatan yang tersedia")
    return

  lihat()
  pilihan = int(input("pilih catatan yang ingin dihapus: "))
  if pilihan < 1 or pilihan > len(catatan):
    print("Nomor catatan tidak valid")
    return

  del catatan[pilihan - 1]
  print("Catatan telah dihapus")

while True:
  print("# ========== Your Note Daily ========= #")
  print("1. Tambah catatan")
  print("2. Lihat catatan")
  print("3. Edit catatan")
  print("4. Hapus cacatan")
  print("5. Keluat")
  choice = int(input("Pilih menu 1 - 5 "))

  if choice == 1:
    tambah()
  elif choice == 2:
    lihat()
  elif choice == 3:
    edit()
  elif choice == 4:
    hapus()
  elif choice == 5:
    print("Arigathanks sudah makai")
    break
  else:
    print("Menu ga ada, pilih 1-5")
```
